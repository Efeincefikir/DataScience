```python
import numpy as np #To use for linear algebra.
import pandas as pd #To use for Data Processing and reading CSV files.
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv") #Contains data for the movie itself.
credits = pd.read_csv("tmdb_5000_credits.csv") #Contains data for the cast behind the movie.
```

```python
movies.head(2) #First 2 entries of the movie dataset.
```

```python
credits.head(2) #First 2 entries of the credits dataset.
```

```python
movies = movies.merge(credits,on='title') #Merging two datasets.
```

```python
movies.head(2) #First two entries of the newly made movies dataset.
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]

#New dataset that includes necessary colums.
```

```python
movies.head(2) #New movies dataset.
```

```python
import ast 
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
#We are simplifying the columns with compicated data with only showing the name part. We will use this in genres and keywords
#columns. We are using ast.literal_eval to treat the data of those columns as dictinary, not string.
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
#Applying convert function to genres and keywords columns.
```

```python
movies.dropna(inplace=True) # deletes null values.
```

```python
movies.head(2)
```

```python
def convertCast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 4:
            L.append(i['name'])
        else: break
        counter+=1
    return L 
#Getting 4 names out of the cast column with the same way we did with genres and keywords.
```

```python
movies['cast'] = movies['cast'].apply(convertCast) #Applying convertCast

```

```python
movies.head()
```

```python
def getDirector(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
#Getting the director from the crew column
```

```python
movies['crew'] = movies['crew'].apply(getDirector) #Applying getDirector
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
#Using this collapse function will make the data cleaner
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
#Applying collapse function
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
#Using this split to make overwiev into a list of words
```

```python
movies.head(3)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
#We will combine every data we tidied into one tag column
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
#Created a dataset with just the title and the tags.
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'Terminator 3: Rise of the Machines'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('The Matrix')
```
